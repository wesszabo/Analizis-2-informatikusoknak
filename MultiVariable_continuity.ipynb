<div class="exercises">
Legyen 
$$
f(x,y) = 
\begin{cases}
\dfrac{x^2y}{x^4+y^2},      & \text{if  } x^2+y^2\neq 0;\\
0,        & \text{if  } x^2+y^2=0.
\end{cases}
$$
Ekkor $f$ folytonos az origóban minden az origón áthaladó egyenes mentén, 
de mint kétváltozós függvény nem folytonos az origóban.
</div>

```
def f(x,y):
    return x**2*y/(x**4+y**2)
```

```
Eq(f(x, m*x), simplify(f(x, m*x)))
```

```
cc = Limit('f(x,mx)', x, 0)
Eq(cc, Limit(f(x,m*x), x, 0).doit())
```

```
print('f(x,x**2) =',f(x,x**2))
```

```
xx = np.linspace(-3,3,100)
yy = np.linspace(-3,3,100)
X, Y = np.meshgrid(xx,yy)
z = f(xx,xx**2)
Z = f(X, Y)
fig = go.Figure(data=[go.Scatter3d(
    x = xx, y = xx**2,
    z = z,
    name = r'$f(x,x^2)$',
    marker = dict(
        size = 0,
    ),
    line = dict(
        color = 'green',
        width = 1
    )
    ),
    go.Scatter3d(
        x = xx, y = xx**2, z = [0 for _ in range(len(xx))],
        name = r'$f(x,x^2)\, \textrm{vetülete } x,y \textrm{ síkon}$',
        marker = dict(size = 0),
        line = dict(
            color = 'gray',
            width = 1
        )
        ),
    go.Surface(
        x=xx, y=yy, z=Z,
        contours_z=dict(show=True, 
                        usecolormap=True,
                        highlightcolor="limegreen", 
                        project_z=True)
        )
    ])
fig.update_layout(title=dict(
                  text=r'$\displaystyle{f(x,y)=\frac{x^2y}{x^4+y^2}}$',
                  font=dict(size=20)),
                  scene_camera_eye=dict(x=1.87, y=0.88, z=0.64),
                  width=700, height=700, 
                  margin=dict(l=65, r=80, b=65, t=70),
                  legend=go.layout.Legend(
        x=0,
        y=1,
        traceorder="normal",
        font=dict(
            family="sans-serif",
            size=16,
            color="black"
        ),
        bgcolor="LightSteelBlue",
        bordercolor="Black",
        borderwidth=2
    )
)
fig.show(renderer="notebook_connected")
```
