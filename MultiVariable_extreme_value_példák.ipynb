```
#@title
def gradient(f, varlist):
    grad = [diff(f, var) for var in varlist]
    return grad        
```

```
##############  JACOBI 
```

```
#@title
def jacobi(funclist, varlist):
    X = Matrix(funclist)
    Y = Matrix(varlist)
    return X.jacobian(Y)
```

## Példa 1.

<div class="exercises">
Legyenek $\mathbf{r}_i\in\mathbf{R}^d$, $i=1,\ldots,n$ adott pontok. Határozzuk meg az 
$$
f(\mathbf{r}):=\sum_{i=1}^n \Vert\mathbf{r}-\mathbf{r}_i  \Vert^2
$$
függvény minimumát.
</div>

<div class="solution">
$$
f'(\mathbf{r})=\sum_{i=1}^n 2(\mathbf{r}-\mathbf{r}_i).
$$
Az $f'(\mathbf{r})=\mathbf{0}$ egyenletből 
$$
\mathbf{r}=\frac{\sum_{i=1}^n \mathbf{r}_i}{n}.
$$
</div>

## Példa 2.

<div class="exercises">
Legyenek $\alpha, \beta, \gamma$ egy háromszög szögei. Határozzuk meg a
$$
g(\alpha,\beta,\gamma):=\sin(\alpha)+\sin(\beta)+\sin(\gamma)
$$
függvény maximumát.
</div>

<div class="solution">
Mivel $\alpha+\beta+\gamma=\pi$ ezért 
$$
g(\alpha,\beta,\gamma)=\sin(\alpha)+\sin(\beta)+\sin(\alpha+\beta)=:f(\alpha,\beta)
$$

</div>

```
#@title
def f(alpha,beta):
    return np.sin(alpha)+np.sin(beta)+np.sin(alpha+beta)
```

```
xx = np.linspace(-0.1,3.2,100)
yy = np.linspace(-0.1,3.2,100)
X, Y = np.meshgrid(xx,yy)
Z = f(X, Y)
fig = go.Figure(data=[
    go.Surface(
        x=xx, y=yy, z=Z,
        contours_z=dict(show=True, 
                        usecolormap=True,
                        highlightcolor="limegreen", 
                        project_z=True)
        )
    ])
fig.update_layout(title=dict(
                  text=r'$f(\alpha,\beta)=\sin(\alpha)+\sin(\beta)+\sin(\alpha+\beta)$',
                  font=dict(size=20)),
                  scene = dict(
                    xaxis_title='α',
                    yaxis_title='β'),
                  scene_camera_eye=dict(x=1.87, y=0.88, z=0.64),
                  width=700, height=700, 
                  margin=dict(l=65, r=80, b=65, t=70),
                  legend=go.layout.Legend(
        x=0,
        y=1,
        traceorder="normal",
        font=dict(
            family="sans-serif",
            size=16,
            color="black"
        ),
        bgcolor="LightSteelBlue",
        bordercolor="Black",
        borderwidth=2
    )
)
fig.show(renderer="notebook_connected")
```

<div class="solution">
Az $f(\alpha,\beta)$ függvény gradiense
</div>

```
func = sin(alpha)+sin(beta)+sin(alpha+beta)
```

```
grad_func = Matrix(gradient(func, [alpha, beta]))
grad_func
```

A

```
Eq1 = Eq(grad_func[0])
Eq2 = Eq(grad_func[1])
Eq1
Eq2
```

egyenletrendszerből adódik
$$
\alpha=\beta.
$$
Visszahelyettesítve, a
$$
\cos{\left (\beta \right )} + \cos{\left (2 \beta \right )}=0
$$
egyenletet kapjuk.

Ezt megoldva kapjuk

```
solveset_real(Eq1.subs({alpha: beta}), beta)
```
